## Scatterplots

```python
# Because there is so much columns, divide them to different chunks first

def pairplots(df):
    hue_col = "class4"
    other_cols = [c for c in df.columns if c != hue_col]

    n_cols = 5
    chunks = math.ceil(len(other_cols)/n_cols)

    for i in range(chunks):
        print(f"Chunk {i}...")
        chunk = other_cols[i*n_cols : (i+1)*n_cols]

        p = sns.pairplot(df[chunk + [hue_col]], hue=hue_col)
        p.savefig(f"./plots/pairplot_{i+1}.png")

# This function was used to generate pairplots which are now saved in a distrinct folder
# Running this code may take some time
# do not uncomment unless want to generate the pics again :3
# pairplots(train)
```
